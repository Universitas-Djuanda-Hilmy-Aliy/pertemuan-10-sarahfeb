Dataset Keras dan Learning Rate

```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
(train_images,train_labels),(test_images,test_labels) = mnist.load_data()
```

```python
print("Ukuran Citra Train",train_images.shape)
print("Jumlah Label Train",train_labels.shape)

print("Ukuran Citra Train",test_images.shape) # 10.000 itu jumlah gambarnya, 28x28 itu ukuran gambar/citranya
print("Jumlah Label Train",test_labels.shape)
```

Menampilkan Data Training

```python
plt.imshow(train_images[50], cmap='gray')
plt.title(train_labels[50])
plt.show()
```

```python
print(set(train_labels))
```

```python
jumlah_baris = 2
jumlah_kolom = 5
jumlah_gambar = 10

gambar, axes = plt.subplots(jumlah_baris, jumlah_kolom, figsize= (1.5*jumlah_kolom, 2*jumlah_baris))

for i in range(jumlah_gambar):
  gambar = axes[i//jumlah_kolom, i%jumlah_kolom]
  gambar.imshow(train_images[i], cmap='gray')
  gambar.set_title('Label = {} '.format(train_labels[i]))
plt.tight_layout()
plt.show()

```

```python
train_images = train_images.reshape((60000,28*28))
train_images = train_images.astype("float32")/255

test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype("float32")/255
```

```python
# Arsitektur Modelnya

import tensorflow as tf
import numpy as np

model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation = 'relu'),
    tf.keras.layers.Dense(10,activation = 'sigmoid')
])
```

```python
model.compile(optimizer ='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_images,train_labels, epochs=100, batch_size=128)
```

batch size/ukuran batch = ini adalah hyperparameter yang menentukan jumlah contoh training yang digunakan dalam 1 iterasi

```python
prediksi = model(test_images)
prediksi = prediksi.numpy() #supaya hasilnya array
prediksi_label = np.argmax(prediksi,axis=1)
prediksi_betul = prediksi_label == test_labels
print(f"akurasi:.{prediksi_betul.mean():.2f}")
```

```python
prediksi_label
```
